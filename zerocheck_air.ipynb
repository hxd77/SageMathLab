```python
F=GF(101)
R.<x1,x2,y1,y2> =F[]

cube=[(0,0),(0,1),(1,0),(1,1)]

def eq(y):
    beta=F(1)
    beta*=x1*y[0]+(1-x1)*(1-y[0])
    beta*=x2*y[1]+(1-x2)*(1-y[1])
    return beta

def mle(values):
    return sum(values[i]*eq(c) for i,c in enumerate(cube))

a=[2,5,11,7]
b=[3,2,8,7]
c=[6,10,88,49]

#我们为a、b、c的值插值一个多项式
A=mle(a)
B=mle(b)
C=mle(c)

# 我们构造约束多项式
# 它在立方体中的所有点处都为零
P=A*B-C
print("P = ",P)

assert P(x1=0,x2=0) == 0
assert P(x1=0,x2=1) == 0
assert P(x1=1,x2=0) == 0
assert P(x1=1,x2=1) == 0

# 我们计算P的MLE
# 由于对于立方体中的所有点，P都是零，因此MLE为零
'''
for cube_point in cube:
    print(cube_point)
    print(cube_point[0],cube_point[1])
'''
P_mle=sum(P(x1=cube_point[0],x2=cube_point[1])*eq([cube_point[0],cube_point[1]])for cube_point in cube)

print("P_mle = ",P_mle)
assert P_mle==0

#在F中选择一个随机点
#我们最后想要证明P_mle(r)=0
r=[29,43] #随机的

#首先，我们计算r处的“相等函数”
eq_r=eq(r)
print("我们计算在r处的相等函数eq_r = ",eq_r)

# 然后我们计算P和eq_r的乘积
S=P*eq_r
print("计算S = P*eq_r = ",S)

#立方体中所有点的S值均为零
assert S(x1=0,x2=0) == 0
assert S(x1=0,x2=1) == 0
assert S(x1=1,x2=0) == 0
assert S(x1=1,x2=1) == 0

#显然，立方体上S的总和为零
H=sum(S(x1=cube_point[0],x2=cube_point[1])for cube_point in cube)
print("立方体S上的总和:",H)
assert H==0

print("===== SUMCHECK =====")
# 求和检验所需的随机性
r_prime=[41,79]

# 记住我们需要限制多项式g1和g2的次数
(g1_deg_bound,g2_deg_bound,*_)=S.degrees() #看每一个变量在所有项中出现的最高次数
print("g1_deg_bound =",g1_deg_bound)
print("g2_deg_bound =",g2_deg_bound)

#第1轮
g1=R(S(x2=0)+S(x2=1))
print("g1 =",g1)
assert g1(x1=0)+g1(x1=1)==H
assert g1.degree()<=g1_deg_bound

#第2轮
g2=R(S(x1=r_prime[0]))
print("g2 = ",g2)
assert g2(x2=0)+g2(x2=1)==g1(x1=r_prime[0])
assert g2.degree()<=g2_deg_bound

print("\nVerifier重新计算S")
print("我们从多项式承诺中接受P(x1=r_prime[0],x2=r_prime[1])")
print("并且Verifier可以轻松自己重新计算eq_r")
S_prime=P(x1=r_prime[0],x2=r_prime[1])*eq_r(x1=r_prime[0],x2=r_prime[1])
#因为P在所有布尔点上都为0，所以只要不管选择哪一个r(\in F),只要x=r相等在布尔超立方体上就为0
print("eq_r(x1=41,x2=79) = ",eq_r(x1=r_prime[0],x2=r_prime[1]))
print("S_prime = ",S_prime)
assert g2(x2=r_prime[1])==S_prime
print("SumCheck 通过")
```
