```python
#欺骗验证者
F=GF(101)
R.<x1,x2> = F[]
Rx.<x> = F[]

P=15*x1*x2+50*x1+11
print(f"P= {P}")

H=P(0,0)+P(0,1)+P(1,0)+P(1,1)
print(f"H = {H}")

g1=P(x1,0)+P(x1,1)
g1=Rx(g1(x,0))
print(f"g1 = {g1}")
assert g1(0)+g1(1)==H

# 现在，让我们尝试欺骗验证者
# 使用拉格朗日插值法伪造g1
fake_g1=Rx.lagrange_polynomial([(0,1),(1,H-1)]) #1+H-1=H
print(f"fake_g1 = {fake_g1}")
print(f"fake_g1!=g1")
#fake_g1通过check
assert fake_g1(1)+fake_g1(0)==H

#Verifier会引入随机数，因此欺骗将在下一轮被发现
r1=71
g2=P(r1,x2)
g2=Rx(g2(0,x))
print(f"Prover定义一个新多项式: g2 = {g2}")
assert g2(0)+g2(1)==g1(r1)

#让我们检查一下这个假的g1是否还能用
print(f"Verifier验证: g2(0)+g2(1) == fake_g1(r1)?",{g2(0)+g2(1) == fake_g1(r1)})

#fake_g1在这一步会失败
assert g2(0)+g2(1)!=fake_g1(r1)

print("\n允许g1是一个更高次的多项式")
#现在，让我们尝试用更高次的多项式来欺骗验证者
points=[(0,1),(1,H-1)]

#我们插值一个多项式，其中该条件在域中的每个点都得到满足
points+=[(i,P(i,0)+P(i,1)) for i in range(2,101)] #所有点值
print(f"points: {points}")
fake_g1=Rx.lagrange_polynomial(points)
print(f"fake_g1: {fake_g1}")
print(f"fake_g1 degree: ",{fake_g1.degree()})

#伪造的g1通过了检查
assert g2(0)+g2(1)==fake_g1(r1)
```
