```python
F=GF(101)
R.<x1,x2> = F[]
P=1+x2+x1+x1*x2

#计算P在布尔超立方体上的总和
H=P(0,0)+P(0,1)+P(1,0)+P(1,1)
print(f"H = {H}")

#得到一个跟P的degree相同的随机多项式Q
#Q=R.random_element(degree=P.degree())
Q=-39*x1^2+44*x1*x2+32*x2^2+24*x1+4*x2
print(f"Q = {Q}")

#计算Q在布尔超立方体上的总和
H_prime=Q(0,0)+Q(0,1)+Q(1,0)+Q(1,1)
print(f"H_prime = {H_prime}")

#随机数rho
#rho=F.random_element()
rho=36
print(f"rho = {rho}")

S=P+rho*Q
print(f"一个新的多项式: S = {S}")
SUM=H+rho*H_prime
print("立方体上S的总和 = H + ρ * H_prime = {SUM}")

#第一轮
print("\nround 1")
g1=S(x2=0)+S(x2=1)
print(f"g1 = {g1}")

#Verifier检查g1是否正确
print(f"g1(0)+g1(1) = {g1(x1=0)+g1(x1=1)}(应该={SUM})")#-26 mod 101=75
assert g1(x1=0)+g1(x1=1)==SUM
assert g1.degree() <= S.degrees()[0] #S.degrees()=(2,2)表示x1和x2的次数
#多项式g的次数必须小于等于i的变量的次数

#Verifier发送一个随机挑战r1
r1=17

#Prover构造g2(x)=P(r1,x)
g2=S(x1=r1)
print(f"g2 = {g2}")

#Verifier检查g2是否正确
print(f"g2(0)+g2(1) = {g2(x2=0)+g2(x2=1)}(应该等于{g1(x1=r1)})")
assert g2(x2=0)+g2(x2=1)==g1(x1=r1)
assert g2.degree()<=S.degrees()[1]

#Verifier发送一个随机挑战r2
r2=70

#最后的验证
assert g2(x2=r2)==S(x1=r1,x2=r2)

#Prover发送H'给Verifier (基于Q的承诺，这本应在一开始就完成的)
#Verifier现在可以取回H
H_final=SUM-rho*H_prime
print(f"H = {H_final}")
```
