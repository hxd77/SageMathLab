```python
p = 7

F7 = GF(p)
R.<u> = F7[]

P = R(u^2+u+3) #不可约多项式
F7_2.<u> = GF(p^2, modulus=P) #二次拓域多项式环

S.<x> = PolynomialRing(F7) #等价于R.<u> = F7[]
G = 4*x^3 + 2*x + 6 #基域多项式

G = G.change_ring(F7_2) #扩域多项式
print("G:", G)

r = F7_2(4*u + 3)
print("result:", G(r))
```

```python
F7 = GF(7)

R.<u> = F7[]
F7_2.<u> = F7.extension(u^2 + u + 3) #使用不可约多项式u^2+u+3作为规则，构建一个二次扩域

S.<v> = F7_2[] #建立一个新的多项式环v

P = S.irreducible_element(3) #找一个最高次幂为3的不可约多项式
print("irreducible poly:", P) 
# v^3 + u*v + 1

F7_6.<v> = F7_2.extension(P) #再进行三次扩域，最终是六次扩域，里面有7^6个元素
```
