```python
p = 7

F7 = GF(p)
R.<u> = F7[]

P = R(u^2+u+3) #不可约多项式
F7_2.<u> = GF(p^2, modulus=P) #二次拓域多项式环

S.<x> = PolynomialRing(F7) #等价于R.<u> = F7[]
G = 4*x^3 + 2*x + 6 #基域多项式

G = G.change_ring(F7_2) #扩域多项式
print("G:", G)

r = F7_2(4*u + 3)
print("result:", G(r))
```

```python
F7 = GF(7)

R.<u> = F7[]
F7_2.<u> = F7.extension(u^2 + u + 3) #使用不可约多项式u^2+u+3作为规则，构建一个二次扩域

S.<v> = F7_2[] #建立一个新的多项式环v
#print(F7_2.random_element())

P = S.irreducible_element(3) #找一个关于变量v的最高次幂为3的不可约多项式，前面的F7_2是多项式的系数
print("不可约多项式P:", P) 
#不可约多项式v^3 + u*v + 1

F7_6.<v> = F7_2.extension(P) #再进行三次扩域，最终是六次扩域，里面有7^6个元素

print(F7_6.random_element())


print(F7_6.random_element())

a=(4*u + 3)*v^2 + (6*u + 5)*v + 4*u + 3
b=(4*u + 1)*v^2 + 5*v + u + 4  

print("a + b = ", a + b)
print("a * b = ", a * b)

'''
a + b =  (u + 4)*v^2 + (6*u + 3)*v + 5*u
a * b =  (3*u + 1)*v^2 + (5*u + 4)*v + 6*u + 5
'''
```

```python
F7=GF(7) #mod 7的有限域
R.<u> =PolynomialRing(F7) #多项式环
F7_2.<u> =F7.extension(u^2+u+3)

S.<v> =PolynomialRing(F7_2)
P=v^3+u*v+1 #S.irreducible_element(3)次数为3的不可约多项式

F7_6.<v> =F7_2.extension(P) 
#这里的元素最高包含v^2，而v前面的系数都是带u的。比如(2u + 6)v^2 + (4u + 3)v + 6。
T.<w> =PolynomialRing(F7_6)

while True:
    Q=T.random_element(degree=2)
    if Q.is_irreducible():
        break
print("不可约多项式Q: ",Q) #Q是一个关于w的二次不可约多项式
# Q:  ((3*u + 4)*v^2 + 3*u*v + 4)*w^2 + ((6*u + 1)*v^2 + (u + 6)*v + u + 6)*w + 6*v^2 + 5*u*v + 3*u + 5

F7_12.<w> =F7_6.extension(Q)
#因为是对 w 的二次扩域，所以最终极的元素最高只有w的一次方，即 Aw + B。而这里的A和B，是一长串带有v和 u的代数式！
print()

c=F7_12.random_element()
d=F7_12.random_element()
print("c -> ",c)
print("d -> ",d)
print("c*d = ",c*d)
```
